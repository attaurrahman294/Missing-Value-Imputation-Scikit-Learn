```python
# now i am going to remove the seles price from the data set becuse this is dependent column 
#in the train data set i am going to assign all independen variable to x and dependent variable to y
x_train = train.drop(columns='SalePrice')
y_test = train['SalePrice']

# now i am going to check the shape of the train data set
print('shape if x_train data',x_train.shape)
print('shape of y_test data',y_test.shape)
```

# numerical missing value imputation 

```python
# i am going to select only the column which has numeric values
num_vars = x_train.select_dtypes(include=["int64","float64"]).columns
num_vars
```

```python
# now i want to check all the 
x_train[num_vars].isnull().sum()/x_train.shape[0]*100
```

```python
# now i am call the funtion SimpleImputer to find mean value in all numerical column
imputer_mean = SimpleImputer(strategy='mean')
# also we can fill constant values in the column where missing value using below script
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)
```

```python
# Now i am put all the mean value in numerical columns where missing values
imputer_mean.fit(x_train[num_vars])
```

```python
# now i have find the below values are the meant value of numerical column 
imputer_mean.statistics_
```

```python
# now we have put all the above value in the numerical column where missing values in list
imputer_mean.transform(x_train[num_vars])
```

```python
x_train[num_vars] = imputer_mean.transform(x_train[num_vars])
#test[num_vars] = imputer_mean.transform(x_train[num_vars])
```

```python
x_train[num_vars].isnull().sum()
```

# Categorical missing value imputation 

```python
# i am going to select only the column which has catagorical values
cat_vars = x_train.select_dtypes(include=["O"]).columns
cat_vars
```

```python
# to check all the catogorical column where missing or null values
x_train[cat_vars].isnull().sum()
```

```python
# now i am call the funtion SimpleImputer to find mean value in all numerical column
imputer_mode = SimpleImputer(strategy='most_frequent')
imputer_mode
# also we can fill constant values in the column where missing value using below script
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)
```

```python

imputer_mode.fit(x_train[cat_vars])
```

```python
imputer_mode.statistics_
```

```python
x_train[cat_vars] = imputer_mode.transform(x_train[cat_vars])
#test[num_vars] = imputer_mean.transform(x_train[num_vars])
x_train.isnull().sum().sum()
```
